# Sampling Distributions

A **sampling distribution** is a probability distribution for all possible values of a statistic computed from a sample of size *n*. The **sampling distribution of the sample mean** is the probability distribution of all possible values of the mean computed from a sample size of *n*. A sampling distributions mean is written as $\bar{x}$. The population mean is written as $\mu$.

## In Class Activity

```python
%matplotlib inline
```

```python
import numpy as np
import itertools as it
from matplotlib import pyplot as plt

def get_sample_means(samples):
    """
    Returns the mean of each sample.
    """
    return [np.mean(sample) for sample in samples]

def get_sampling_mean(samples):
    """
    Returns the sampling distribution of the sample mean for the given samples.
    """
    return np.mean(get_sample_means(samples))

def get_samples(population, sample_size):
    """
    Returns all possible samples from the population at sample size n.
    """
    return list(it.combinations(population, sample_size))

def draw_hist(sample_means, sample_size):
    hist, bins = np.histogram(sample_means, 5)
    hist = hist / float(len(sample_means))
    width = 0.7*(bins[1]-bins[0])
    center = (bins[:-1]+bins[1:])/2
    plt.title("Sampling distribution at n=%(sample_size)d" % {"sample_size": sample_size})
    plt.bar(center, hist, align = 'center', width = width)
    plt.show()

six_students = [np.random.normal(50, 10) for i in range(6)]

for sample_size in range(2, 6):
    samples = get_samples(six_students, sample_size)
    sample_means = get_sample_means(samples)
    draw_hist(sample_means, sample_size)
    sampling_mean = get_sampling_mean(sample_means)
    print "Mean:", np.mean(six_students)
    print "Sampling Mean:", sampling_mean
```

I have noted in the above that as *n* increases, the distribution is squished together. So as *n* increases the standard deviation of the sampling distribution decreases.

## A Larger Sampling Distribution

```python
# The Standford-Binet IQ test is normally distributed with a mean score
# of 100 and a standard deviation of 15. Below I'm creating the sampling
# distribution for this dataset at n=9.

n = 9
mean = 100
std_dev = 15

samples = [np.random.normal(mean, std_dev, n) for i in range(1000)]
sample_means = get_sample_means(samples)
hist, bins = np.histogram(sample_means, 10)
width = 0.7*(bins[1]-bins[0])
center = (bins[:-1]+bins[1:])/2
plt.title("Sampling distribution at n=9")
plt.bar(center, hist, align = 'center', width = width)
plt.show()

print "Sampling distribution standard deviation:", np.std(sample_means)

```

I observe that the sampling distribution looks like a normal distributions. I also observe that the standard deviation has gotten smaller. Meanwhile, the mean of the distribution hasn't really changed.

## The Mean And Standard Deviation of The Sampling Distribution

As we have seen the standard deviation goes down as the sample size goes up. It turns out that the standard deviation is always going to be $\alpha_\bar{x} = \frac{\alpha}{\sqrt{n}}$. This is also called the **standard error of the mean**. Meanwhile, the mean, as the graphs I've created have shown, doesn't really change. The shape of the ditribution should also be normal when drawing from a normal distribution.

## Combining Sampling Distributions With Z-Scores

```python
# We are still going to use the distribution from the last problem.
# However, this time the sample size will be 10. The question we are
# asking is this: P(sample mean > 110) or what is the probability that
# the sample mean is greater then 110.

from scipy.stats import norm

sample_mean = mean
sample_size = 10
sample_std_dev = std_dev / np.sqrt(sample_size)
z_score = (110 - sample_mean) / sample_std_dev
print "Z-score:", z_score
print "P(sample mean > 110):", 1 - norm.cdf(z_score)

```

# What Happens When The Underlying Distribution Isn't Normal

The sampling distribution is still normal. Especially as the size of the sample is increased. This is called the Central Limit Theorem. Even for something like dice, this theorem applies. As the sample size increases things tend to be closer to the average. How far they are from it is the part that makes things look normal. Its probably also why the standard deviation was called the standard error of the mean.
